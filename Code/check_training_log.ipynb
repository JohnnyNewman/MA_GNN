```python
df_results = pd.concat([pd.read_excel(f"training_log/{fn}") for fn in os.listdir("training_log") if fn.endswith(".xlsx")]).sort_values("data_dir").sort_values("epoch")
```

```python
plt.scatter(df_results["loss"], df_results["val_loss"], c=df_results["epoch"])
```

```python
C = np.cov(df_results[["loss", "val_loss", "epoch"]].T)
C
```

```python
for i in range(3):
    for j in range(i):
        r = C[j,i] / np.sqrt(C[i,i]*C[j,j])
        print(j, i, r, r**2)
```

```python
df = df_results.groupby("epoch").mean()

for l in ['loss', 'loss_data', 'loss_bc', 'val_loss']:
    plt.plot(df.index, np.log10(df[l]), label=l)
    # plt.plot(df.index, df[l], label=l)

plt.legend()
```

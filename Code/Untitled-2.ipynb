```python
from scipy import sparse
import scipy.sparse.linalg
```

```python
mesh.cell_type(2)
```

```python
A = sparse.dok_matrix((n_points, n_points))

for cell_type in [5,9]:
    cells = np.hstack((mesh.cells_dict[cell_type][:], mesh.cells_dict[cell_type][:,:1]))
    for cell in cells:
        for i,j in zip(cell[:-1], cell[1:]):
            A[i,j] = 1
            A[j,i] = 1

D = A.sum(axis=0).squeeze()

L = A.copy()

for i in range(n_points):
    L[i,i] = -D[0,i]
```

```python
A = A.tocsr()
```

```python
L = L.tocsr()
```

```python
s,v = sparse.linalg.eigs(L, k=10)
```
